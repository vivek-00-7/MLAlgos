####Handling Outliers

#####Detecting Outliers Using BoxPlot

```python
# Plotting box plots for all numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(10, 6))
df[numerical_columns].boxplot()
plt.title('Box Plot for Numerical Columns')
plt.show()

# Identifying columns with outliers
outliers_columns = []

for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    if df[(df[col] < lower_bound) | (df[col] > upper_bound)].any(axis=None):
        outliers_columns.append(col)

print("\nColumns with outliers:", outliers_columns)
```

#####Handling Outliers Using Winsorization

```python
# Applying capping (winsorizing) to columns with outliers
for col in outliers_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap the outliers
    df[col] = np.where(df[col]>upper_bound,upper_bound, np.where(df[col]<lower_bound,lower_bound,df[col]))
```

```python
# Plotting box plots for all numerical columns after capping
for col in outliers_columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot for {col} after Capping Outliers')
    plt.show()
```

####Encoding Categorical Data(Categorical->Numerical)

Step - 1: Identify your type of Categorical Data
- Nominal : There aren't any order or relationship between those categories.<br>Example:State(Telangana,TamilNadu etc) , Branch(CSE,ECE,EEE,IT etc).
- Ordinal : There is an order or relationship between the categories.<br>
Example:Grade(S,A,B,C,D) , Review(Excellent,Good,Okay,Bad).

Step - 2:Encoding Technique
- Nominal : One Hot Encoding
- Ordinal : Ordinal Encoding
<br>
Note:If target column is categorical(usually we see in multiclass classification) , then we use LABEL ENCODER THERE INSTEAD OF ONE HOT OR ORDINAL.

```python
df = pd.read_csv('/content/customer.csv')
df.sample(5)
```

- Nominal Variables - Gender , Purchased
- Ordinal Variables - Review , education

#####OrdinalEncoding(for ordinal columns)

```python
from sklearn.preprocessing import OrdinalEncoder
# Defining the mappings for ordinal encoding -give in the order
review_mapping = ['Poor', 'Average', 'Good'] #poor-0,avg-1,good-2
education_mapping = ['School', 'UG', 'PG'] #school-0,ug-1,pg-2

# Initializing the OrdinalEncoder with the specified mappings
ordinal_encoder = OrdinalEncoder(categories=[review_mapping, education_mapping])

# Encoding 'review' and 'education' columns
df[['review', 'education']] = ordinal_encoder.fit_transform(df[['review', 'education']])

# Converting the float results to integers
df[['review', 'education']] = df[['review', 'education']].astype(int)
print(df.head())
```

#####Label Encoder (for target Column)

```python
from sklearn.preprocessing import  LabelEncoder
# Initializing the LabelEncoder for the target column
label_encoder = LabelEncoder()

# Encoding the 'purchased' column
df['purchased'] = label_encoder.fit_transform(df['purchased'])

# Display the resulting DataFrame
print(df.head())
```

#####One Hot Encoding(for nominal columns)

Can be implemented in two ways
- Pandas getdummies()
- sklearn OneHotEncoder (Better to use)

```python
# Performing one-hot encoding on the 'gender' column
df = pd.get_dummies(df, columns=['gender'], drop_first=True)
df.sample(5)
#Applies one-hot encoding to the 'gender' column using pd.get_dummies,
# with drop_first=True to avoid the dummy variable trap
#(which drops one of the one-hot encoded columns to avoid multicollinearity).
```

```python
from sklearn.preprocessing import OneHotEncoder
# Identifying the nominal columns to be one-hot encoded
nominal_columns = ['gender', 'city']

# Initializing the OneHotEncoder
onehot_encoder = OneHotEncoder(drop='first', sparse=False)

# Performing one-hot encoding on the nominal columns
onehot_encoded = onehot_encoder.fit_transform(df[nominal_columns])

# Creating a DataFrame for the one-hot encoded columns
onehot_encoded_df = pd.DataFrame(onehot_encoded, columns=onehot_encoder.get_feature_names_out(nominal_columns))

# Dropping the original nominal columns and concatenating the one-hot encoded columns
df = df.drop(nominal_columns, axis=1)
df = pd.concat([df, onehot_encoded_df], axis=1)
```

### Ensemble Methods

#### Voting Classifier

The idea behind the VotingClassifier is to combine conceptually different machine learning classifiers and use a majority vote or the average predicted probabilities (soft vote) to predict the class labels. Such a classifier can be useful for a set of equally well performing models in order to balance out their individual weaknesses.



```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
clf1 = DecisionTreeClassifier()
clf2 = LogisticRegression(random_state=1,solver='liblinear',max_iter=1000)
clf3 = BernoulliNB()
vclf1 = VotingClassifier(estimators=[('dt',clf1),('lr',clf2),('nb',clf3)],voting='hard')
vclf2 = VotingClassifier(estimators=[('dt',clf1),('lr',clf2),('nb',clf3)],voting='soft')
```

```python
print("Hard Voting Results:")
for clf,label in zip([clf1,clf2,clf3,vclf1],['Decision Tree','Logistic Regression','Bernoulli Naive Bayes','Voting Classifier']):
  scores = cross_val_score(clf,X_train,y_train,scoring='accuracy',cv=5)
  print("%0.2f [%s]"%(scores.mean(),label))
print("Soft Voting Results:")
for clf,label in zip([clf1,clf2,clf3,vclf1],['Decision Tree','Logistic Regression','Bernoulli Naive Bayes','Voting Classifier']):
  scores = cross_val_score(clf,X_train,y_train,scoring='accuracy',cv=5)
  print("%0.2f [%s]"%(scores.mean(),label))
```

Decision Tree alone performs better than voting classifier , hence it is optimal to use decision trees for classification.

####Bagging *Classifier*

In ensemble algorithms, bagging methods form a class of algorithms which build several instances of a black-box estimator on random subsets of the original training set and then aggregate their individual predictions to form a final prediction. These methods are used as a way to reduce the variance of a base estimator (e.g., a decision tree), by introducing randomization into its construction procedure and then making an ensemble out of it.

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
scores_knn = cross_val_score(knn,X_train,y_train,scoring='accuracy',cv=5)
print("accuracy of single KNN Classifier:",scores_knn.mean())
bagging = BaggingClassifier(KNeighborsClassifier())
scores_bagging = cross_val_score(bagging,X_train,y_train,scoring='accuracy',cv=5)
print("accuracy of Bagging Classifier:",scores_bagging.mean())
```

<b>RandomForest Classifier:</b> individual decision trees typically exhibit high variance and tend to overfit. The injected randomness in forests yield decision trees with somewhat decoupled prediction errors. By taking an average of those predictions, some errors can cancel out. Random forests achieve a reduced variance by combining diverse trees, sometimes at the cost of a slight increase in bias. In practice the variance reduction is often significant hence yielding an overall better model.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=None, min_samples_split=2,random_state=0)
scores_dt = cross_val_score(dt,X_train,y_train,cv=5)
print("accuracy of a single Decision Tree Classifier:",scores_dt.mean())
rf = RandomForestClassifier(n_estimators = 100,max_depth=None)
scores_rf = cross_val_score(rf,X_train,y_train,cv=5)
print("accuracy of Random Forest Classifier:",scores_rf.mean())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
X = df.drop('country',axis=1)

# Scale the features using MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
df = pd.DataFrame(X_scaled,columns=X.columns)
df
```

```python
#Elbow Plot
wcss = []
max_clusters = 10  # maximum number of clusters to try
for i in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plotting the elbow method
plt.figure(figsize=(10, 5))
plt.plot(range(1, max_clusters + 1), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.xticks(range(1, max_clusters + 1))
plt.grid(True)
plt.show()
```

```python
 #Perform KMeans clustering with a chosen number of clusters (let's say 5)
kmeans = KMeans(n_clusters=3)
y_kmeans = kmeans.fit_predict(df)
y_kmeans
```

```python
df['cluster'] = y_kmeans
df
```

```python
# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(df[df['cluster']==0]['income'], df[df['cluster']==0]['life_expec'], s=50, c='red', label='Cluster 1')
plt.scatter(df[df['cluster']==1]['income'], df[df['cluster']==1]['life_expec'], s=50, c='blue', label='Cluster 2')
plt.scatter(df[df['cluster']==2]['income'], df[df['cluster']==2]['life_expec'], s=50, c='green', label='Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='black', marker='*',
            label='Centroids')
plt.title('KMeans Clustering')
plt.xlabel('Income (k$)')
plt.ylabel('Life expectancy')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
max_clusters = 10  # maximum number of clusters to try
for i in range(2, max_clusters + 1):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df)
    silhouette_scores.append(silhouette_score(df, kmeans.labels_))

# Plot silhouette scores
plt.figure(figsize=(10, 5))
plt.plot(range(2, max_clusters + 1), silhouette_scores, marker='o', linestyle='--')
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage

```

```python
# Hierarchical clustering with single linkage
plt.figure(figsize=(10, 6))
dendrogram(linkage(df, method='single'), leaf_rotation=90, leaf_font_size=8)
plt.title('Dendrogram (Single Linkage)')
plt.show()
```

```python
# Hierarchical clustering with Ward linkage
plt.figure(figsize=(10, 6))
dendrogram(linkage(df, method='ward'), leaf_rotation=90, leaf_font_size=8)
plt.title('Dendrogram (Ward Linkage)')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering

max_clusters = 10
clusters_range = range(2, max_clusters + 1)

# Define linkage methods
linkage_methods = ['single', 'complete', 'average', 'ward']

# Create a dataframe to store the results
results_df = pd.DataFrame(index=clusters_range, columns=linkage_methods)

# Calculate metrics for each linkage method and each number of clusters
for method in linkage_methods:
    for n_clusters in clusters_range:
        # Perform hierarchical clustering
        agglomerative = AgglomerativeClustering(n_clusters=n_clusters, linkage=method)
        cluster_labels = agglomerative.fit_predict(df)

        # Calculate silhouette score
        silhouette_avg = silhouette_score(df, cluster_labels)

        # Store silhouette score in the results dataframe
        results_df.at[n_clusters, method] = silhouette_avg

# Convert all values to float
results_df = results_df.astype(float)

# Plot the comparison table
plt.figure(figsize=(10, 6))
sns.heatmap(results_df, annot=True, cmap="YlGnBu", fmt=".3f")
plt.title('Silhouette Scores Comparison')
plt.xlabel('Linkage Method')
plt.ylabel('Number of Clusters')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
